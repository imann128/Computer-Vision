```python
from ultralytics import YOLO
# Loading the model
model = YOLO("yolo11n.pt")
results = model('/content/cars.png', save = True, show = True)
results[0].show()
```

```python
import pandas as pd
# Saving data in a list
data = []
for result in results: # Looping through result
    for box in result.boxes:
        data.append([box.cls.item(), box.conf.item(), box.xyxy.tolist()]) # Saving in a nested list
# Making a dataframe
df = pd.DataFrame(data, columns=["Class", "Confidence", "Bounding Box"])
# Saving to a CSV file
df.to_csv("runs/detect/predict/detections.csv", index=False)
# Downloading
from google.colab import files
files.download("runs/detect/predict/detections.csv")
```
